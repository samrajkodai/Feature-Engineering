```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
%matplotlib.inline
```

```python
df=pd.read_csv("Advertising.csv",index_col=0)
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
df.head()
```

```python
x=sm.add_constant(x)
model=sm.OLS(y,x).fit()
```

```python
model
```

```python
model.summary()
```

```python
x.corr()
```
